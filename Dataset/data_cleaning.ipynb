```python
import pandas as pd
import numpy as np
df = pd.read_csv('SAMA_Cleaned (2).csv')
print(df.info())
```

```python
df['period'] = pd.to_datetime(df['period'])
df = df.fillna(0)
```

```python
df['local_pct'] = (df['local_assets_total'] / df['total_assets'] * 100).round(2)
df['foreign_pct'] = (df['foreign_assets_total'] / df['total_assets'] * 100).round(2)
df = df.replace([np.inf, -np.inf], 0).fillna(0)
```
